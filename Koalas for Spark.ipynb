```
loan_data=ks.read_csv('dbfs:/FileStore/tables/Q4_2018.csv')
```

```
ks.set_option('display.max_rows',50)
```

```
ld_sel=loan_data[['loan_status','loan_amnt','grade','term']]
```

```
ld_sel['loan_status'].value_counts()
```

```
ld_sel.groupby('grade').agg({'grade':'count'})
```

```
ld_sel.groupby('grade').agg({'grade':'count'})/ld_sel['grade'].count()
```

```
ld_tmp=ks.concat([ld_sel.groupby('grade').agg(grade_count=('grade','count')),ld_sel.groupby('grade').agg(grade_per=('grade','count'))/ld_sel['grade'].count()],axis=1)
```

```
ld_tmp.head()
```

```
ld_tmp.reset_index()
```

```
ks.sql("select * from {ld_tmp} where grade_count >250")
```

```
ld_tmp.sort_index()
```

```
ld_tmp.sort_values(by='grade_count')
```
